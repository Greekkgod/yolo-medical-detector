```python
# Cell 3: Real-time Webcam Detection
model = YOLO('yolov8n.pt')
cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read()
    results = model(frame, device=0)  # GPU inference
    annotated = results[0].plot()
    cv2.imshow('Medical YOLO RTX 3050', annotated)
    if cv2.waitKey(1) == ord('q'): break

cap.release()
cv2.destroyAllWindows()
```
